```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("supermarket_sales.csv")

# Calculate Total_Sale using loop
total_sales = []
for i in range(len(data)):
    total = data["Quantity"][i] * data["Unit_Price"][i]
    total_sales.append(total)

data["Total_Sale"] = total_sales

# Show the data
print("Supermarket Sales Data with Total Sales:")
print(data)

# Total Revenue
total_revenue = 0
for sale in data["Total_Sale"]:
    total_revenue += sale
print("\nTotal Revenue:", total_revenue)

# Revenue per Product
product_revenue = {}
for i in range(len(data)):
    product = data["Product"][i]
    total = data["Total_Sale"][i]
    if product in product_revenue:
        product_revenue[product] += total
    else:
        product_revenue[product] = total

print("\nRevenue per Product:")
for product, revenue in product_revenue.items():
    print(product, ":", revenue)

# Plot Revenue per Product (Bar Chart)
plt.figure(figsize=(10, 5))
plt.bar(product_revenue.keys(), product_revenue.values(), color='orange')
plt.title("Revenue per Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Best Selling Product by Quantity
product_quantity = {}
for i in range(len(data)):
    product = data["Product"][i]
    quantity = data["Quantity"][i]
    if product in product_quantity:
        product_quantity[product] += quantity
    else:
        product_quantity[product] = quantity

best_product = ""
max_quantity = 0
for product, quantity in product_quantity.items():
    if quantity > max_quantity:
        max_quantity = quantity
        best_product = product

print("\nBest Selling Product (by quantity):", best_product)

# Revenue by Category
category_revenue = {}
for i in range(len(data)):
    category = data["Category"][i]
    total = data["Total_Sale"][i]
    if category in category_revenue:
        category_revenue[category] += total
    else:
        category_revenue[category] = total

print("\nRevenue by Category:")
for category, revenue in category_revenue.items():
    print(category, ":", revenue)

# Plot Revenue by Category (Pie Chart)
plt.figure(figsize=(6, 6))
plt.pie(category_revenue.values(), labels=category_revenue.keys(), autopct='%1.1f%%', startangle=140)
plt.title("Revenue by Category")
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle.
plt.tight_layout()
plt.show()
```
